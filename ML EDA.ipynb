13. What is the average depreciation in selling price based on the bike's age (current year - manufacturing
year)

```python
import pandas as pd

# Load dataset
file_path = "/mnt/BIKE DETAILS.csv"  # Update file path if necessary
df = pd.read_csv(file_path)

# Define the current year
CURRENT_YEAR = 2024

# Check if required columns exist
if "year" in df.columns and "selling_price" in df.columns:
    # Calculate bike age
    df["bike_age"] = CURRENT_YEAR - df["year"]

    # Remove bikes with zero age (to avoid division by zero)
    df = df[df["bike_age"] > 0]

    # Calculate average depreciation per year
    df["depreciation_per_year"] = df["selling_price"] / df["bike_age"]

    # Compute the overall average depreciation
    avg_depreciation = df["depreciation_per_year"].mean()

    print(f"📉 **Average Depreciation Per Year:** ₹{avg_depreciation:,.2f}")

else:
    print("⚠️ The dataset does not contain 'year' or 'selling_price' columns. Check df.columns to verify.")
```

14.  Which bike names are priced significantly above the average price for their manufacturing year?

```python
import pandas as pd

# Load dataset
file_path = "/mnt/BIKE DETAILS.csv"  # Update if necessary
df = pd.read_csv(file_path)

# Check if required columns exist
if "year" in df.columns and "selling_price" in df.columns and "name" in df.columns:
    # Compute mean and standard deviation of selling price per year
    year_stats = df.groupby("year")["selling_price"].agg(["mean", "std"]).reset_index()

    # Merge with original dataframe
    df = df.merge(year_stats, on="year", how="left")

    # Define threshold: Above mean + 1.5 * standard deviation
    df["threshold"] = df["mean"] + (1.5 * df["std"])

    # Identify bikes priced significantly above average
    high_priced_bikes = df[df["selling_price"] > df["threshold"]][["name", "year", "selling_price"]]

    # Display results
    print("🏍️ **Bikes Priced Significantly Above Average for Their Year:**")
    print(high_priced_bikes)

else:
    print("⚠️ The dataset is missing 'year', 'selling_price', or 'name' columns. Check df.columns to verify.")
```

9.  How do car prices vary based on the customer’s annual income bracket?

14.  What is the correlation between car price, engine size, and annual income of customers, and how do these
features interact?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Load dataset
file_path = "/mnt/Car Sale.csv"
df = pd.read_csv(file_path)

# Check column names to avoid KeyErrors
print("Dataset Columns:", df.columns)

# Drop rows with missing values in relevant columns
df = df.dropna(subset=["Price ($)", "Engine", "Annual Income"])

# Extract numeric engine size from "Engine" column
def extract_engine_size(engine):
    match = re.search(r'\d+', str(engine))
    return int(match.group()) if match else None

df["Engine_Size"] = df["Engine"].apply(extract_engine_size)
df = df.dropna(subset=["Engine_Size"])  # Drop invalid engine size values

# Convert data to numeric
df["Price ($)"] = pd.to_numeric(df["Price ($)"], errors="coerce")
df["Annual Income"] = pd.to_numeric(df["Annual Income"], errors="coerce")

# Compute correlation matrix
correlation_matrix = df[["Price ($)", "Engine_Size", "Annual Income"]].corr()

# Display correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Price, Engine Size & Annual Income")
plt.show()

# Pairplot to visualize interactions
sns.pairplot(df[["Price ($)", "Engine_Size", "Annual Income"]], diag_kind="kde")
plt.suptitle("Pairwise Relationships: Price, Engine Size & Annual Income", y=1.02)
plt.show()

# 3D Scatter Plot for Interaction
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(df["Engine_Size"], df["Annual Income"], df["Price ($)"], c=df["Price ($)"], cmap="coolwarm", alpha=0.7)

# Labels and Titles
ax.set_xlabel("Engine Size")
ax.set_ylabel("Annual Income")
ax.set_zlabel("Price ($)")
ax.set_title("3D Scatter: Engine Size, Income & Price")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/mnt/spotify.csv"
df = pd.read_csv(file_path)

# Check if required columns exist
if "Popularity" in df.columns and "Duration (ms)" in df.columns:
    # Convert columns to numeric
    df["Popularity"] = pd.to_numeric(df["Popularity"], errors="coerce")
    df["Duration (ms)"] = pd.to_numeric(df["Duration (ms)"], errors="coerce")

    # Drop missing values
    df_cleaned = df.dropna(subset=["Popularity", "Duration (ms)"])

    # Convert duration from milliseconds to minutes for better readability
    df_cleaned["Duration_min"] = df_cleaned["Duration (ms)"] / 60000

    # Scatter plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df_cleaned["Duration_min"], y=df_cleaned["Popularity"], alpha=0.5, color="blue")
    plt.xlabel("Track Duration (Minutes)")
    plt.ylabel("Popularity Score")
    plt.title("Relationship Between Track Duration and Popularity")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

else:
    print("⚠️ The dataset does not contain 'popularity' or 'duration_ms' columns.")
```

4.  Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using
a countplot?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/mnt/spotify.csv"  # Ensure the file path is correct
df = pd.read_csv(file_path)

# Check if 'Artist' column exists
if "Artist" in df.columns:
    # Count the number of tracks per artist
    artist_counts = df["Artist"].value_counts().reset_index()
    artist_counts.columns = ["Artist", "Track Count"]

    # Get the top 10 artists with the most tracks
    top_artists = artist_counts.head(10)

    # Plot the countplot without the deprecated warning
    plt.figure(figsize=(12, 6))
    sns.barplot(x="Track Count", y="Artist", data=top_artists, hue="Artist", dodge=False, legend=False)
    plt.xlabel("Number of Tracks")
    plt.ylabel("Artist")
    plt.title("Top 10 Artists with the Most Tracks")
    plt.show()

    # Display the artist with the most tracks
    most_tracks_artist = artist_counts.iloc[0]
    print(f"🎵 The artist with the most tracks is '{most_tracks_artist['Artist']}' with {most_tracks_artist['Track Count']} tracks.")

else:
    print("⚠️ The dataset does not contain an 'Artist' column.")
```

6. Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and
display the average popularity for each artist?

```python
import pandas as pd

# Load the dataset
file_path = "/mnt/spotify.csv"  # Ensure the correct file path
df = pd.read_csv(file_path)

# Check if the required columns exist
required_columns = ["Artist", "Popularity"]
if all(col in df.columns for col in required_columns):
    # Convert 'Popularity' to numeric if necessary
    df["Popularity"] = pd.to_numeric(df["Popularity"], errors="coerce")

    # Calculate the average popularity for each artist
    artist_avg_popularity = df.groupby("Artist")["Popularity"].mean().reset_index()

    # Sort artists by average popularity in descending order
    top_5_artists = artist_avg_popularity.sort_values(by="Popularity", ascending=False).head(5)

    # Identify the artist with the highest average popularity
    most_popular_artist = top_5_artists.iloc[0]

    # Display results
    print("🎤 **Top 5 Most Popular Artists (by Average Popularity)** 🎤")
    print(top_5_artists)

    print(f"\n🏆 The artist with the **highest** average popularity is **{most_popular_artist['Artist']}** "
          f"with an average popularity score of **{most_popular_artist['Popularity']:.2f}**.")

else:
    print(f"⚠️ The dataset is missing one or more required columns: {required_columns}")
```

7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist

```python
import pandas as pd

# 1. Get Top 5 Artists
top_artists = df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False).head(5).index.tolist()

# 2. Get Most Popular Tracks for Each Artist
for artist in top_artists:
    top_track = df[df['Artist'] == artist].sort_values(by='Popularity', ascending=False).iloc[0]['Track Name']
    print(f"Top track for {artist}: {top_track}")
```

8. Visualize relationships between multiple numerical variables simultaneously using a pair plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/mnt/spotify.csv"  # Ensure the correct file path
df = pd.read_csv(file_path)

# Selecting numerical columns for visualization
numeric_columns = ["Popularity", "Duration_ms", "Danceability", "Energy", "Loudness", "Tempo"]

# Check if the required numerical columns exist
existing_columns = [col for col in numeric_columns if col in df.columns]

if existing_columns:
    # Create a pair plot
    sns.pairplot(df[existing_columns], diag_kind="kde", corner=True)

    # Show the plot
    plt.suptitle("Pair Plot of Numerical Features in Spotify Dataset", y=1.02)
    plt.show()
else:
    print(f"⚠️ The dataset is missing numerical columns from: {numeric_columns}")
```

9.  Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
violin plot?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/mnt/spotify.csv"
df = pd.read_csv(file_path)

# Check if required columns exist
if "Artist" in df.columns and "Duration (ms)" in df.columns:

    # Convert 'Duration (ms)' to numeric (handle errors)
    df["Duration (ms)"] = pd.to_numeric(df["Duration (ms)"], errors="coerce")

    # Select the top 10 artists with the most tracks
    top_artists = df["Artist"].value_counts().head(10).index
    df_filtered = df[df["Artist"].isin(top_artists)]

    # Set figure size
    plt.figure(figsize=(12, 6))

    # Violin Plot (Fixed)
    sns.violinplot(x="Artist", y="Duration (ms)", data=df_filtered, inner="quartile", hue="Artist", legend=False)

    # Customize plot
    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Artist")
    plt.ylabel("Duration (ms)")
    plt.title("Distribution of Track Durations Across Top 10 Artists")
    plt.tight_layout()

    plt.show()

else:
    print("⚠️ The dataset is missing the 'Artist' or 'Duration (ms)' column.")
```

10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
violin plot?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/mnt/spotify.csv"
df = pd.read_csv(file_path)

# Check if required columns exist
if "Artist" in df.columns and "Popularity" in df.columns:

    # Convert 'Popularity' to numeric (handle errors)
    df["Popularity"] = pd.to_numeric(df["Popularity"], errors="coerce")

    # Select the top 10 artists with the most tracks
    top_artists = df["Artist"].value_counts().head(10).index
    df_filtered = df[df["Artist"].isin(top_artists)]

    # Set figure size
    plt.figure(figsize=(12, 6))

    # Violin Plot (Fixed)
    sns.violinplot(x="Artist", y="Popularity", data=df_filtered, inner="quartile", hue="Artist", legend=False)

    # Customize plot
    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Artist")
    plt.ylabel("Popularity")
    plt.title("Distribution of Track Popularity Across Top 10 Artists")
    plt.tight_layout()

    plt.show()

    # Swarm Plot (Fixed)
    plt.figure(figsize=(12, 6))
    sns.swarmplot(x="Artist", y="Popularity", data=df_filtered, hue="Artist", legend=False, size=4)

    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Artist")
    plt.ylabel("Popularity")
    plt.title("Swarm Plot of Track Popularity Across Top 10 Artists")
    plt.tight_layout()

    plt.show()

else:
    print("⚠️ The dataset is missing the 'Artist' or 'Popularity' column.")
```
